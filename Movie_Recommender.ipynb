```python
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.shape
```

```python
movies.dtypes
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
#ast is used to convert string into list
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i["name"])
  return l
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies["genres"].head()
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies["cast"][0]
```

```python
def convert2(obj):
  l=[]
  cnt=0
  for i in ast.literal_eval(obj):
    if cnt!=3:
      l.append(i["name"])
      cnt+=1
    else:
      break
  return l
```

```python
movies["cast"]=movies["cast"].apply(convert2)
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
def crew_func(obj):
  l=[]
  for i in ast.literal_eval(obj):
   if i["job"]=="Director":
     l.append(i["name"])
     break;
  return l
```

```python
movies["crew"]=movies["crew"].apply(crew_func)
```

```python
movies["overview"][0]
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
 movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["overview"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
New_df=movies[["movie_id","title","tags"]]
```

```python
New_df
```

```python
#tags column is a list,converting it into string
New_df["tags"]=New_df["tags"].apply(lambda x:" ".join(x))
New_df.head()
```

```python
New_df["tags"][0]
```

```python
#converting everything in lowercase
New_df["tags"]=New_df["tags"].apply(lambda x:x.lower())
```

```python
New_df.head()
```

```python
#DOING STEMING
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
New_df["tags"]=New_df["tags"].apply(stem)
```

```python
New_df["tags"][1]
```

**Vectorization(using bag of words)**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(New_df["tags"]).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
#calculating cosine distance between vectors
from sklearn.metrics.pairwise import cosine_similarity
#similarity is inverse of distance
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
#MAIN FUNCTION FOR RECOMMENDING MOVIES
def recommend(movie):
  movie_index=New_df[New_df["title"]==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(New_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
